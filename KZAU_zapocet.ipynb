Regulátory jsou klíčovou součástí moderních řídicích systémů, které najdou uplatnění v široké škále průmyslových i spotřebitelských aplikací. Slouží k udržení požadované hodnoty výstupní veličiny systému tím, že dynamicky reagují na jakékoli změny v systému nebo vnějším prostředí. Existuje několik základních typů regulátorů, z nichž každý má své specifické vlastnosti a použití.

##Výhody použití regulátorů
- Umožňují automatické řízení a stabilizaci regulované veličiny
- Zlepšují kvalitu a přesnost regulačního procesu
- Reagují na poruchy a vlivy vnějšího prostředí
- Zvyšují efektivitu a výkonnost regulovaného systému

##Metody návrhu regulátorů
- Frekvenční metody (Nyquistova a Nichollsova křivka)
- Kořenová metoda (metoda kořenových trajektorií)
- Optimalizační metody (záporná zpětná vazba, minimální integrál regulační odchylky)
- Experimentální metody (Ziegler-Nicholsonova metoda)

#Typy regulátorů
##P (Proporcionální) regulátory
- Princip: Výstupní signál regulátoru je proporcionální rozdíl mezi požadovanou hodnotou (setpoint) a aktuální hodnotou měřené veličiny.

- Diferenciální rovnice:
- Kp = proporcionální zesílení
- e(t) regulační odchylka.
$$
u(t) = K_p \cdot e(t)
$$

##I Regulátor (Integrační)
- Princip: Výstupní signál je proporcionální časovému integrálu regulační odchylky, což umožňuje eliminaci stálé regulační odchylky.
- Diferenciální rovnice:
- Ki = integrační zesílení
$$
u(t) = K_i \int e(t) \, dt
$$

##D Regulátor (Derivační)
- Princip: Výstupní signál je proporcionální derivaci regulační odchylky podle času, reaguje na rychlost změny odchylky.
- Diferenciální rovnice:
- Kd = derivační zesílení
$$
u(t) = K_d \frac{de(t)}{dt}
$$

##PI Regulátor
- Kombinuje principy P a I regulátorů pro dosažení rychlé reakce na změny a eliminaci stálé regulační odchylky.
-Diferenciální rovnice:
$$
u(t) = K_p \cdot e(t) + K_i \int e(t) \, dt
$$

##PD Regulátor
- Kombinuje P a D regulátory pro rychlou reakci na změny bez překmitu způsobeného integrační složkou.
- Diferenciální rovnice:
$$
u(t) = K_p \cdot e(t) + K_d \frac{de(t)}{dt}
$$

##PID Regulátor
- Nejvíce používaný typ regulátoru, kombinuje výhody P, I, a D regulátorů.
- Diferenciální rovnice:
$$
u(t) = K_p \cdot e(t) + K_i \int e(t) \, dt + K_d \frac{de(t)}{dt}
$$

##Úvod do frekvenčních charakteristik
Frekvenční charakteristiky jsou grafické reprezentace, jak systém reaguje na různé frekvenční vstupy. Zahrnují dvě základní charakteristiky:

- Amplitudová charakteristika: Ukazuje, jak se mění amplituda výstupního signálu systému v závislosti na frekvenci vstupního signálu.
- Fázová charakteristika: Ukazuje, jak se mění fázový posun mezi vstupním a výstupním signálem systému v závislosti na frekvenci.

Tyto charakteristiky jsou klíčové pro analýzu dynamiky systému, umožňují nám porozumět, jak systém reaguje na změny vstupního signálu různých frekvencí, což je zásadní pro návrh efektivního a stabilního řídicího systému.

##Význam frekvenčních charakteristik
Frekvenční charakteristiky hrají zásadní roli v procesu návrhu řídicích systémů, protože:

- Umožňují analyzovat stabilitu systému. Například, kritéria stability Nyquistova nebo Bodeho kritérium mohou být použita k určení, zda je systém stabilní nebo ne.
- Poskytují informace potřebné k optimalizaci výkonu systému, včetně rychlosti odezvy, přesnosti a odolnosti proti rušení.
- Pomáhají při identifikaci rezonančních frekvencí a jiných klíčových vlastností systému, které mohou ovlivnit jeho chování.

##Analýza frekvenčních charakteristik
K analýze frekvenčních charakteristik se běžně používají následující nástroje:

- Bodeho diagram: Umožňuje grafické znázornění amplitudové a fázové charakteristiky systému jako funkci frekvence. Bodeho diagram je široce používán pro návrh a analýzu systémů zpětnovazebního řízení.

- Nyquistův diagram: Zobrazuje komplexní frekvenční odezvu systému v komplexní rovině. Tento diagram je užitečný pro posouzení stability systému pomocí Nyquistova kritéria stability.

- Nicholsův diagram: Kombinuje amplitudovou a fázovou charakteristiku v jednom grafu. Používá se pro přímý návrh zpětnovazebních smyček s využitím specifických kontur zisku a fáze.

Tyto nástroje umožňují inženýrům a designérům systémů lépe porozumět chování systému a efektivněji navrhovat řídicí strategie pro dosažení požadované stability a výkonu.

Upevněte znalosti pomocí doplňovačky níže.

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Otázky a správné odpovědi
questions = {
    "Frekvenční charakteristiky jsou grafické reprezentace, jak systém reaguje na různé _________ vstupy.": "frekvenční",
    "_________ charakteristika ukazuje, jak se mění amplituda výstupního signálu systému.": "Amplitudová",
    "_________ charakteristika ukazuje, jak se mění fázový posun mezi vstupem a výstupem.": "Fázová",
    "Bodeho diagram je široce používán pro _________ a analýzu systémů zpětnovazebního řízení.": "návrh",
}

# Vytvoření widgetů pro otázky a tlačítko pro kontrolu
answer_widgets = [widgets.Text(description=f"Otázka{i+1}:") for i in range(len(questions))]
check_button = widgets.Button(description="Kontrola odpovědí")
output = widgets.Output()

# Funkce pro kontrolu odpovědí
def check_answers(b):
    with output:
        clear_output()
        correct = 0
        for answer_widget, correct_answer in zip(answer_widgets, questions.values()):
            if answer_widget.value.lower().strip() == correct_answer.lower():
                correct += 1
        if correct == len(questions):
            print("Všechny odpovědi jsou správné! Skvělá práce!")
        else:
            print(f"{correct}/{len(questions)} odpovědí jsou správné. Zkuste to znovu.")

# Registrace události pro tlačítko
check_button.on_click(check_answers)

# Zobrazení widgetů
for question, answer_widget in zip(questions.keys(), answer_widgets):
    display(widgets.Label(value=question))
    display(answer_widget)
display(check_button, output)
```

##Úvod do přechodových charakteristik
###Definice a význam přechodových charakteristik v kontextu řídicích systémů

Přechodové charakteristiky jsou grafické reprezentace odezvy systému na určité vstupní signály v čase. Tyto charakteristiky nám umožňují vizualizovat a analyzovat, jak systém reaguje na změny, a jsou klíčové pro pochopení dynamického chování systémů.

###Dva základní typy přechodových charakteristik, které se běžně analyzují, jsou:

- Skoková odezva: Odezva systému na skokový vstup, tedy na náhlou změnu vstupního signálu z nulové hodnoty na konstantní hodnotu.
- Impulsní odezva: Odezva systému na krátkodobý impuls, což je vstupní signál velmi krátké doby trvání a vysoké amplitudy.
Význam přechodových charakteristik

Přechodové charakteristiky poskytují cenné informace o stabilitě, rychlosti reakce a přesnosti systému při sledování požadovaných hodnot. Analýza těchto charakteristik pomáhá inženýrům při návrhu a ladění řídicích systémů tak, aby splňovaly specifické požadavky na chování a výkon.

##Analýza a interpretace přechodových charakteristik
###Klíčové parametry přechodových charakteristik

- Čas vzestupu: Doba, která uplyne od aplikace vstupního signálu do okamžiku, kdy výstup poprvé dosáhne požadované hodnoty.
- Překmit: Maximální procentuální odchylka výstupu od jeho konečné hodnoty v průběhu přechodového děje.
- Čas ustálení: Doba potřebná k tomu, aby se výstup ustálil v blízkosti konečné hodnoty, obvykle definováno jako doba, po které se výstup již neodchyluje od konečné hodnoty o více než určité procento.
Stálá stavová chyba: Rozdíl mezi požadovanou hodnotou (setpoint) a skutečnou konečnou hodnotou výstupu systému.
Vliv nastavení regulátoru

Nastavení parametrů regulátoru (např. Kp, Kd, Ki u PID regulátoru) má výrazný vliv na přechodové charakteristiky systému. Správným výběrem těchto parametrů lze:

- Zkrátit čas ustálení, což znamená, že systém rychleji dosáhne požadovaného stavu.
- Minimalizovat nebo úplně eliminovat překmit, což vede k stabilnějšímu chování systému.
- Snížit nebo eliminovat stálou stavovou chybu, což zlepšuje přesnost systému v dosažení a udržení požadované hodnoty.

Grafy níže znázorňují, jak vypadá příliš vysoké Kp, Ki a Kd.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Funkce pro modelování dynamiky systému s PID regulátorem
def pid_system(y, t, Kp, Ki, Kd, setpoint):
    error = setpoint - y
    # Pro tento příklad použijeme jednoduchou integraci a derivaci chyby
    integral_error = error * t
    derivative_error = error / t if t > 0 else 0
    u = Kp*error + Ki*integral_error + Kd*derivative_error
    dydt = u  # Zjednodušený model systému, kde výstup PID ovlivňuje přímo dydt
    return dydt

# Inicializační parametry
t = np.linspace(0, 2, 100)  # časová osa
y0 = 0  # počáteční podmínka
setpoint = 1  # cílová hodnota skokové funkce

# Parametry PID pro jednotlivé scénáře
scenarios = {
    "Vysoké Kp": (30, 0, 0),
    "Vysoké Ki": (0, 50, 0),
    "Vysoké Kd": (0, 0, 5)
}

plt.figure(figsize=(18, 6))

# Vykreslení grafů pro každý scénář
for i, (title, (Kp, Ki, Kd)) in enumerate(scenarios.items(), 1):
    y = odeint(pid_system, y0, t, args=(Kp, Ki, Kd, setpoint))
    plt.subplot(1, 3, i)
    plt.plot(t, y, label=f"Odezva systému\nKp={Kp}, Ki={Ki}, Kd={Kd}")
    plt.plot(t, np.full_like(t, setpoint), 'r--', label='Setpoint')
    plt.title(title)
    plt.xlabel('Čas')
    plt.ylabel('Výstupní hodnota')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```

Díky cvičení níže si můžete simulovat regulátory i včetně konstant a hodnot.

```python

import ipywidgets as widgets
from ipywidgets import interactive
import numpy as np
import matplotlib.pyplot as plt

# Funkce pro simulaci regulátoru
def simulate_regulator(regulator_type, Kp, Ki, Kd, time_end=20):
    time = np.linspace(0, time_end, 1000)
    setpoint = 1
    y = 0
    e_sum = 0
    e_prev = 0

    y_values = []
    for t in time:
        e = setpoint - y
        e_sum += e
        if regulator_type == "P":
            u = Kp * e
        elif regulator_type == "I":
            u = Ki * e_sum
        elif regulator_type == "D":
            u_d = Kd * (e - e_prev)
            e_prev = e
            u = u_d
        elif regulator_type == "PI":
            u = Kp * e + Ki * e_sum
        elif regulator_type == "PD":
            u_d = Kd * (e - e_prev)
            e_prev = e
            u = Kp * e + u_d
        elif regulator_type == "PID":
            u_d = Kd * (e - e_prev)
            e_prev = e
            u = Kp * e + Ki * e_sum + u_d
        y += 0.6 * u
        y_values.append(y)

    plt.figure(figsize=(10, 6))
    plt.plot(time, y_values, label='Output')
    plt.axhline(setpoint, color='r', linestyle='--', label='Setpoint')
    plt.xlabel('Čas')
    plt.ylabel('Regulovaná veličina')
    plt.title(f'Simulace {regulator_type} regulátoru')
    plt.legend()
    plt.show()

# Widgety
regulator_type_widget = widgets.RadioButtons(options=['P', 'I', 'D', 'PI', 'PD', 'PID'], description='Regulátor:')
Kp_widget = widgets.FloatSlider(value=1.0, min=0, max=5, step=0.1, description='Kp:')
Ki_widget = widgets.FloatSlider(value=0.2, min=0, max=2, step=0.1, description='Ki:')
Kd_widget = widgets.FloatSlider(value=0.5, min=0, max=5, step=0.1, description='Kd:')

# Interaktivní widgety spojené s funkcí simulace
interactive_plot = interactive(simulate_regulator, {'manual': True, 'manual_name': 'Simulovat'}, regulator_type=regulator_type_widget, Kp=Kp_widget, Ki=Ki_widget, Kd=Kd_widget)
display(interactive_plot)
```
